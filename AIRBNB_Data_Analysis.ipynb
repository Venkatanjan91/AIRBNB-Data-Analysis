```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('airbnbdata.csv')
```

```python
df.head()
```

```python
df.columns
```

**Checking Missing Values**

```python
print(df.isnull().sum())
```

# Handling Missing Values

```python
df["last review"] = pd.to_datetime(df["last review"], errors = "coerce")
```

```python
df.info()
```

```python
df.fillna({"reviews per month" : 0, "last review" : df["last review"].min()}, inplace = True)
```

```python
df.dropna(subset=["NAME", "host name"], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns = ["license","house_rules"], errors = "ignore")
```

```python
df.head()
```

```python
#Remove Dollar signs and convert into float
df["price"] = df["price"].replace("[\$,]", "", regex=True).astype(float)
df["service fee"] = df["service fee"].replace("[\$,]", "", regex=True).astype(float)
```

```python
df.head()
```

# Remove Duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
# descriptive statistics
df.describe()
```

# Visualization

**What is the distribution of listing prices?**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True, color='blue')
plt.title('Distribution of Listing Prices')
plt.xlabel('Price $')
plt.ylabel('Frequency')
plt.show()
```

**How are different room types distributed?**

```python
df["room type"]
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='room type', palette='viridis')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

**How are listings distributed across the neighbourhood?**

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='neighbourhood group', color='green', order=df['neighbourhood group'].value_counts().index)
plt.title('Neighbourhood Distribution')
plt.ylabel('Neighbourhood Group')
plt.xlabel('Count')
plt.show()
```

**What is the relationship between price and room type?**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='room type', y='price', palette='viridis')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price $')
plt.legend(title='Room Type')
plt.show()
```

**How has the number of reviews change over time?**

```python
df["last review"] = pd.to_datetime(df["last review"])
reviews_overtime = df.groupby(df["last review"].dt.to_period("M")).size()

plt.figure(figsize=(12, 6))
reviews_overtime.plot(kind='line', color='purple')
plt.title('Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
```
